```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import accuracy_score,classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('/kaggle/input/creditcard-fraud-detection/creditcard.csv')
data.head()
```

```python
data[['Amount','Time','Class']].describe()
```

```python
# this will show all the rows so its hard to find the null value
data.isna()
# this will return the true value if any value is missing in the whole column
data.isna().any()
```

```python
null_values = pd.DataFrame({'Columns':data.isna().sum().index,'No. of null values':data.isna().sum().values,'Percentage':data.isna().sum().values/data.shape[0]})
null_values
```

```python
nfcount = 0
notfraud = data['Class']
len(notfraud)
```

```python
for i in range(len(notfraud)):
    if notfraud[i]==0:
        nfcount+=1
nfcount
```

```python
per_nf = (nfcount/len(notfraud)) * 100
print("persentage of not fraud or safe transaction in the data : ",per_nf)
```

```python
f_count = len(notfraud) - nfcount
per_f = 100 - per_nf
print('persentage of fraud transaction in the data : ',per_f)
f_count
```

```python
plot_data = pd.DataFrame()
plot_data['Fraud'] = [f_count]
plot_data['Genuine'] =[ nfcount]
plot_data
```

```python
sns.barplot( data = plot_data)
plt.ylabel('Total Transaction')
plt.title("Bar graph for fraud & genuine transactions")
plt.show()
```

```python
plt.title("count plot for fraud & genuine transactions")
sns.countplot(x = "Class",data = data)
plt.show()
```

```python
x = data.drop('Class',axis = 1)
y = data['Class']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
logR = LogisticRegression()
logR.fit(x_train,y_train)
```

```python
y_pred = logR.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```
